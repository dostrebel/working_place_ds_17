# Imports

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# URL Lesen

```python
url = "https://www.iaaf.org/records/toplists/sprints/100-metres/outdoor/men/senior/2017"
response = requests.get(url)
sprinters_soup = BeautifulSoup(response.text, 'html.parser')
```

# Headers auslesen

```python
headers = []
for elem in sprinterliste.find_all('th'):
    headers.append(elem.text.strip())
```

```python
headers
```

# List Comprehension (For loop abgek√ºrzt)

```python
[x for x in headers if x != '']
```

# Namen heraus filtern

```python
namens_list = sprinterliste.find_all('td', {'data-th':'Competitor'})
```

```python
type(namens_list)
```

```python
len(namens_list)
```

```python
namen = []
for elem in namens_list:
    namen.append(elem.text.strip())
```

```python
namen
```

# IDs aus HREF links extrahieren

```python
id_list = sprinterliste.find_all('td', {'data-th':'Competitor'})
```

```python
len(id_list)
```

```python
ids = []
for elem in id_list:
    ids.append(elem.find('a')['href'].replace('/athletes/athlete=', ''))
```

```python
ids
```
